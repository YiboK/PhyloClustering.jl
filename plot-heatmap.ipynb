```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math   
```

```python
df = pd.read_csv("./data/kmeans_4taxa.csv")
df.index = df.index + 1
plt.figure(figsize=(12,12))
plot = sns.heatmap(df, annot=True, cmap="crest",vmin = 0, vmax = 1);
plot.xaxis.tick_top()
plot.tick_params(axis='y', rotation=0)
```

```python
# function to plot heatmap for 15 trees with their changed trees
def plot_heatmap(df_list):
    fig, axes = plt.subplots(5, 3, figsize=(20, 16))
    col = 0
    row = 0
    for i in range(0,15):
        plot = sns.heatmap(df_list[i], annot=True, cmap="crest",vmin = 0, vmax = 1, ax=axes[math.floor(i/3), i - 3*math.floor(i/3)]);
        plot.xaxis.tick_top()
        plot.tick_params(axis='y', rotation=0)
```

```python
change = ["org","inc","inc_r","dec","dec_r","all_r"]
dfs = []
for i in range(1,16):
    df = pd.read_csv("./data/kmeans_4_tree_" + str(i) + ".csv")
    df.columns = change
    df.index = change
    dfs.append(df)
plot_heatmap(dfs)
```

```python
dfs = []
for i in range(1,16):
    df = pd.read_csv("./data/GMM_prob_4_tree_" + str(i) + ".csv")
    df.columns = change
    df.index = change
    dfs.append(df)
plot_heatmap(dfs)
```

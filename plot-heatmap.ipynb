```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math   
```

```python
# function to plot heatmap for 15 trees with their changed trees
def plot_heatmap(df_list, trees, title = ''):
    fig, ax = plt.subplots(5, 3, figsize=(18, 16))
    col = 0
    row = 0
    for i in range(0,15):
        plot = sns.heatmap(df_list[i], annot=True, cmap="crest",vmin = 0, vmax = 1, 
                           ax=ax[math.floor(i/3), i - 3*math.floor(i/3)], cbar = False);
        plot.xaxis.tick_top()
        plot.set_title(trees[i])
        plot.tick_params(axis='y', rotation=0)
    fig.suptitle(title, fontsize='xx-large')
    plt.tight_layout()
```

```python
trees = ["(((4,3),2),1)","(((4,2),3),1)","(((3,2),4),1)","(((4,3),1),2)","(((4,1),3),2)",
         "(((1,3),4),2)","(((1,2),4),3)","(((4,2),1),3)","(((4,1),2),3)","(((2,1),3),4)",
         "(((1,3),2),4)","(((3,2),1),4)","((4,3),(2,1))","((4,1),(2,3))","((4,2),(3,1))"]
```

```python
df = pd.read_csv("./data/kmeans_4taxa_100.csv")
df.columns = trees
df.index = trees

plt.figure(figsize=(12,12))
plt.title('100 trees K-means')
plot = sns.heatmap(df, annot=True, cmap="crest",vmin = 0, vmax = 1);
plot.xaxis.tick_top()
plot.tick_params(axis='x', rotation=45)
plot.tick_params(axis='y', rotation=0)

df = pd.read_csv("./data/GMM_prob_4taxa_100.csv")
df.columns = trees
df.index = trees

plt.figure(figsize=(12,12))
plt.title('100 trees GMM posterior probability')
plot = sns.heatmap(df, annot=True, cmap="crest",vmin = 0, vmax = 1);
plot.xaxis.tick_top()
plot.tick_params(axis='x', rotation=45)
plot.tick_params(axis='y', rotation=0)

df = pd.read_csv("./data/GMM_llpg_4taxa_100.csv")
df.columns = trees
df.index = trees

plt.figure(figsize=(12,12))
plt.title('100 trees GMM log likelihood')
plot = sns.heatmap(df, annot=True, cmap="crest",vmin = 0, vmax = 1);
plot.xaxis.tick_top()
plot.tick_params(axis='x', rotation=45)
plot.tick_params(axis='y', rotation=0)
```

```python
from itertools import combinations

result_kmeans = pd.read_csv("./data/kmeans_4taxa_5000.csv")
result_kmeans = result_kmeans.to_numpy()
result_GMM = pd.read_csv("./data/GMM_prob_4taxa_5000.csv")
result_GMM = result_GMM.to_numpy()
all_comb = list(combinations(range(0,15),2))
good_kmeans = []
good_GMM = []
good_both = []
bad_both = []
for comb in all_comb:
    i = comb[0]
    j = comb[1]
    pair = (i+1,j+1)
    if result_GMM[i,j] > 0.7 and result_kmeans[i,j]>0.7:
        good_both.append(pair)
        good_kmeans.append(pair)
        good_GMM.append(pair)
    elif result_kmeans[i,j] > 0.7:
        good_kmeans.append(pair)
    elif result_GMM[i,j] > 0.7:
        good_GMM.append(pair)
    else:
        bad_both.append(pair)
```

```python
print("Both good: ", good_both)
print("Kmeans good: ", good_kmeans)
print("GMM good: ", good_GMM)
print("Both bad: ", bad_both)
```

```julia
using RCall
using PhyloNetworks
@rimport SiPhyNetwork
```

```julia
R"set.seed(123)"
rcopy(R"tree = SiPhyNetwork::sim.bdh.taxa.ssa(8, 15, 1, 0, 0, 0)") ;
```

```julia
trees = []
for i in 1:15
    tree = rcopy(R"SiPhyNetwork::write.net(tree[$i], digits = 1)")
    tree = readTopology(tree)
    for e in tree.edge
        e.length = 0
    end
    print("case$i = \""writeTopology(tree)*"\"\n")
end
```
